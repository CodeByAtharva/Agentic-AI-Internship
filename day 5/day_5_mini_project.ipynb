###Travel Itinerary Builder using LLM (PromptTemplate + Chain)

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata


city=input("Enter your City: ")
days=float(input("Enter the number of days: "))
style=input("Enter your Travel Style: ")
interest=input("Enter your Interest: ")

llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",google_api_key=userdata.get('Gemini_key'))

bmi_prompt=PromptTemplate(template="""you are a Travel Itinerary Builder, who has extensive experience in creating a daywise planners,
you should do the following:
1) Create a Day-wise plan
Must generate the plan as:

Day 1
Day 2
… up to the number of days entered by the user
2) Include time blocks for every day
Each day must contain:

Morning
Afternoon
Evening
3) Add one budget-saving tip per day
At the end of each day, Gemini must add:

✅ Budget Tip: (one practical money-saving suggestion),

the output format is(strict):
you must return output exactly in this structure:

Day 1

Morning: ...
Afternoon: ...
Evening: ...
Budget Tip: ...
Day 2

Morning: ...
Afternoon: ...
Evening: ...
Budget Tip: ...
(continue till all days),

input:
City: {city},
Number of days: {days},
Travel style: {style},
Interest: {interest}
""",input_variables=["city","days","style","interest"])

planner_chain=bmi_prompt | llm | StrOutputParser()
print(planner_chain.invoke({"city":city,"days":days,"style":style,"interest":interest}))
```
