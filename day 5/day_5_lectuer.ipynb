##Lanchain using PromptTemplate


```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from google.colab import userdata

lang="java"
n=1

prompt=f"Generate a joke on following programming language {text}"
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",google_api_key=userdata.get('Gemini_key'))

temp=PromptTemplate(template="Generate a {n} joke on following programming language {lang}", input_variables=["lang","n"])

chain=temp | llm

res=chain.invoke({"lang":lang,"n":n})

print(res)
```

##Langchain output parser


```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata

lang="java"
n=1

prompt=f"Generate a joke on following programming language {text}"
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",google_api_key=userdata.get('Gemini_key'))

temp=PromptTemplate(template="Generate a {n} joke on following programming language {lang}", input_variables=["lang","n"])

chain=temp | llm | StrOutputParser() # add this function here it will get a string out of the llm

res=chain.invoke({"lang":lang,"n":n})

print(res)
```

#Multiple chain

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata

topic="Quantum Computing"

prompt=f"Generate a joke on following programming language {text}"
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",google_api_key=userdata.get('Gemini_key'))

content_temp=PromptTemplate(template="Give me 5 line detailed content on {topic}", input_variables=["topic"])

summary_temp=PromptTemplate(template="Give me 5 point short summary on {a}", input_variables=["a"])

chain=content_temp | llm | StrOutputParser() | summary_temp | llm | StrOutputParser()

res=chain.invoke({"topic":topic})

print(res)
```

##Topic from user -> generate 5 interview questions -> ideal answers -> followup question

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata

topic="Quantum Computing"

prompt=f"Generate a joke on following programming language {text}"
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",google_api_key=userdata.get('Gemini_key'))

question_prompt=PromptTemplate(template="Generate 5 interview question on {topic}", input_variables=["topic"])
answer_prompt=PromptTemplate(template="Give me it's ideal answers {question}", input_variables=["question"])
followup_prompt=PromptTemplate(template="give me followup questions {answer}", input_variables=["answer"])


chain=content_temp | llm | StrOutputParser() | summary_temp | llm | StrOutputParser()
interview_chain= question_prompt | llm | StrOutputParser() | answer_prompt | llm | StrOutputParser() | followup_prompt | llm | StrOutputParser()

res=interview_chain.invoke({"topic":topic})

print(res)
```
