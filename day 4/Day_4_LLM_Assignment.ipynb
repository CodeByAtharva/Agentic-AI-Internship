##Question 5

```python
from google import genai
from google.colab import userdata

def generate_response(prompt):
  res = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents=prompt,
)
  return res.text

topic = input("Enter your topic: ")
audience = input("Enter your audience: ")
length = input("Enter your length: ")
tone = input("Enter your tone: ")

prompt1=f"""
topic: {topic},
audience: {audience}
Task: Generate a clear outline with exactly 5 section titles.
    Return ONLY the titles as a bulleted list
"""
outline_res=generate_response(prompt1)

prompt2=f"""
Using ONLY the following outline: {outline_res}
    Write a piece of content with these constraints:
    - Audience: {audience}
    - Tone: {tone}
    - Target Length: {length}
    - Order: Follow the outline sequence exactly.,
    return a json format
"""

content_res=generate_response(prompt2)

print(content_res)
```
