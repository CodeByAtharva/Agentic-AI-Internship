##Langchain Parallel

```python
text="Unemployement in India"

def get_word_count(text):
  return len(text.split())

def get_char_count(text):
  return len(text)
print(f"Word count: {get_word_count(text)}")
print(f"Char count: {get_char_count(text)}")
```

```python
from langchain_core.runnables import RunnableLambda, RunnableParallel
wc= RunnableLambda(get_word_count)
cc= RunnableLambda(get_char_count)

parallel= RunnableParallel({
    "word":wc,
    "char":cc
})
res=parallel.invoke(text)
res
```
